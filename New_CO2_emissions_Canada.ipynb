The dataset titled "CO2 Emissions in Canada" contain information on various vehicle models and their corresponding CO2 (carbon dioxide emissions), fuel consumption metrics in Canada. The dataset consists of 7,385 entries and is structured with the following 12 attributes:

- Make: The brand of the vehicle.
- Model: The specific model of the vehicle.
- Vehicle Class: The category of the vehicle (e.g., SUV, compact).
- Engine Size (L): The volume of the engine in liters.
- Cylinders: The number of cylinders in the vehicle's engine.
- Transmission: The type of transmission (e.g., automatic, manual).
- Fuel Type: The type of fuel the vehicle uses (e.g., gasoline, diesel).
- Fuel Consumption City (L/100 km): The fuel consumption in urban areas measured in liters per 100 kilometers.
- Fuel Consumption Hwy (L/100 km): The fuel consumption on highways measured in liters per 100 kilometers.
- Fuel Consumption Comb (L/100 km): The combined fuel consumption (city and highway) measured in liters per 100 kilometers.
- Fuel Consumption Comb (mpg): The combined fuel consumption (city and highway) measured in miles per gallon.
- CO2 Emissions(g/km): The CO2 emissions of the vehicle measured in grams per kilometer.

This dataset can provide insights into the environmental impact of different vehicle types, and consumers in making informed decisions regarding vehicle purchases and usage based on environmental considerations. It also serves as a resource for studying trends in vehicle emissions, evaluating the efficiency of different engine types and fuel systems, and advancing initiatives towards reducing vehicular emissions. â€‹

**1. CO2 Emissions Analysis:**

```python
# 1.1 What is the distribution of CO2 emissions across all vehicles?

plt.figure(figsize=(10, 6))
plt.hist(df['CO2 Emissions(g/km)'], bins=50, color='blue', alpha=0.7)
plt.title('Distribution of CO2 Emissions Across All Vehicles')
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
co2_distribution = df['CO2 Emissions(g/km)'].describe()
co2_distribution
```

* The distribution of CO2 emissions is roughly unimodal, with the most frequent values concentrated in a specific range.
* There is a peak around the middle of the distribution, suggesting that a significant number of vehicles have CO2 emissions in that range.
* The frequency decreases as the CO2 emissions values move towards the higher end, indicating fewer vehicles with very high emissions.
* The tail of the distribution extends towards the higher CO2 emissions values, showing that while they are less frequent, there are still some vehicles with very high CO2 emissions.


```python
# 1.2. How do CO2 emissions vary by vehicle class?

co2_by_class = df.groupby('Vehicle Class')['CO2 Emissions(g/km)'].mean().sort_values()

plt.figure(figsize=(12, 8))
co2_by_class.plot(kind='barh', color='skyblue')
plt.title('Average CO2 Emissions by Vehicle Class')
plt.xlabel('Average CO2 Emissions (g/km)')
plt.ylabel('Vehicle Class')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
co2_by_class = df.groupby('Vehicle Class')['CO2 Emissions(g/km)'].mean().sort_values()
co2_by_class
```

* The bar chart illustrates the average CO2 emissions by vehicle class. Each bar represents the average emissions of a specific vehicle class, ordered from the class with the lowest average emissions at the bottom to the class with the highest average emissions at the top.

* This visualization helps identify which vehicle classes, on average, are associated with higher or lower CO2 emissions. It can be a valuable insight for policy-making, consumer choice, and understanding the impact of different vehicle types on the environment.


```python
# 1.3. Is there a correlation between engine size (in liters) and CO2 emissions?

correlation = df['Engine Size(L)'].corr(df['CO2 Emissions(g/km)'])

plt.figure(figsize=(10, 6))
plt.scatter(df['Engine Size(L)'], df['CO2 Emissions(g/km)'], alpha=0.5, color='green')
plt.title(f'Engine Size vs. CO2 Emissions (Correlation: {correlation:.2f})')
plt.xlabel('Engine Size (L)')
plt.ylabel('CO2 Emissions (g/km)')
plt.grid(True)
plt.show()
```

```python
correlation_engine_co2 = df['Engine Size(L)'].corr(df['CO2 Emissions(g/km)'])
correlation_engine_co2
```

* The scatter plot visualizes the relationship between engine size (in liters) and CO2 emissions (in grams per kilometer), with a calculated correlation coefficient of approximately 0.85.

* This strong positive correlation coefficient indicates that there is a significant direct relationship between engine size and CO2 emissions. As engine size increases, CO2 emissions tend to increase as well.

```python
# 1.4. Which car Make has the highest impact on CO2 emissions?

co2_impact_by_make = df.groupby('Make')['CO2 Emissions(g/km)'].mean()

highest_co2_impact_make = co2_impact_by_make.idxmax()
least_co2_impact_make = co2_impact_by_make.idxmin()
highest_co2_impact_value = co2_impact_by_make.max()
least_co2_impact_value = co2_impact_by_make.min()

print("Make with the highest CO2 emissions:", highest_co2_impact_make, "with an average of", highest_co2_impact_value, "g/km")
print("Make with the least CO2 emissions:", least_co2_impact_make, "with an average of", least_co2_impact_value, "g/km")
```

```python
avg_co2_by_make = df.groupby('Make')['CO2 Emissions(g/km)'].mean()
avg_co2_by_make = avg_co2_by_make.sort_values(ascending=False)

plt.figure(figsize=(10, 10))
sns.barplot(x=avg_co2_by_make.values, y=avg_co2_by_make.index)
plt.title('Average CO2 Impact by Vehicle Make')
plt.xlabel('Vehicle Make')
plt.ylabel('Average CO2 Impact')
plt.xticks(rotation=90)
plt.show()
```

**2. Fuel Consumption Insights:**

```python
# 2.1 How does fuel consumption (city, highway, and combined) vary across different vehicle classes?

fuel_consumption_by_class = df.groupby('Vehicle Class').agg({
    'Fuel Consumption City (L/100 km)': 'mean',
    'Fuel Consumption Hwy (L/100 km)': 'mean',
    'Fuel Consumption Comb (L/100 km)': 'mean'
}).sort_values(by='Fuel Consumption Comb (L/100 km)')

fig, ax = plt.subplots(figsize=(12, 8))

fuel_consumption_by_class.plot(kind='barh', ax=ax, colormap='viridis')
ax.set_title('Average Fuel Consumption by Vehicle Class')
ax.set_xlabel('Fuel Consumption (L/100 km)')
ax.set_ylabel('Vehicle Class')
ax.grid(axis='x')
plt.tight_layout()
plt.show()
```

* The bar chart shows the average fuel consumption in liters per 100 kilometers for city driving, highway driving, and combined (average of city and highway) across different vehicle classes. Each vehicle class is represented by a horizontal set of three bars, each bar color-coded to represent a type of fuel consumption.

* From this visualization, we can compare how each vehicle class performs in different driving conditions. Typically, vehicles have higher fuel consumption in the city than on the highway due to frequent stops and starts, which require more fuel. The combined average gives a more general view of a vehicle's overall fuel efficiency across driving conditions.

```python
# 2.2 Which vehicle make and model are the most and least fuel-efficient?

fuel_efficiency = df.groupby(['Make', 'Model'])['Fuel Consumption Comb (L/100 km)'].mean()

most_fuel_efficient = fuel_efficiency.idxmin()
least_fuel_efficient = fuel_efficiency.idxmax()
most_fuel_efficient_value = fuel_efficiency.min()
least_fuel_efficient_value = fuel_efficiency.max()

most_fuel_efficient, most_fuel_efficient_value, least_fuel_efficient, least_fuel_efficient_value
```

```python
fuel_efficiency = pd.DataFrame({
    'Make': ['HYUNDAI', 'CHEVROLET'],
    'Model': ['IONIQ BLUE', 'EXPRESS 3500 PASSENGER FFV'],
    'Fuel Consumption Comb (L/100 km)': [most_fuel_efficient_value, least_fuel_efficient_value]
})

plt.figure(figsize=(10, 4))
ax = fuel_efficiency.plot(kind='barh', x='Model', y='Fuel Consumption Comb (L/100 km)',
                                   legend=False, color=['green', 'red'])
ax.set_title('Most vs. Least Fuel-Efficient Vehicles')
ax.set_xlabel('Fuel Consumption (Combined L/100 km)')
ax.set_ylabel('Vehicle Make and Model')
ax.set_xlim(0, max(fuel_efficiency['Fuel Consumption Comb (L/100 km)']) + 5)
plt.tight_layout()
plt.show()
```

* The most fuel-efficient vehicle make and model in the dataset is the Hyundai Ioniq Blue, with an average combined fuel consumption of 4.1 liters per 100 kilometers. On the other end, the least fuel-efficient vehicle is the Chevrolet Express 3500 Passenger FFV, with an average combined fuel consumption of 22.55 liters per 100 kilometers.

```python
# 2.3 Is there a correlation between fuel consumption and CO2 emissions?

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# City fuel consumption vs. CO2 emissions
axes[0].scatter(df['Fuel Consumption City (L/100 km)'], df['CO2 Emissions(g/km)'], color='blue', alpha=0.5)
axes[0].set_title('City Fuel Consumption vs. CO2 Emissions')
axes[0].set_xlabel('City Fuel Consumption (L/100 km)')
axes[0].set_ylabel('CO2 Emissions (g/km)')

# Highway fuel consumption vs. CO2 emissions
axes[1].scatter(df['Fuel Consumption Hwy (L/100 km)'], df['CO2 Emissions(g/km)'], color='green', alpha=0.5)
axes[1].set_title('Highway Fuel Consumption vs. CO2 Emissions')
axes[1].set_xlabel('Highway Fuel Consumption (L/100 km)')
axes[1].set_ylabel('CO2 Emissions (g/km)')

# Combined fuel consumption vs. CO2 emissions
axes[2].scatter(df['Fuel Consumption Comb (L/100 km)'], df['CO2 Emissions(g/km)'], color='red', alpha=0.5)
axes[2].set_title('Combined Fuel Consumption vs. CO2 Emissions')
axes[2].set_xlabel('Combined Fuel Consumption (L/100 km)')
axes[2].set_ylabel('CO2 Emissions (g/km)')

plt.tight_layout()
plt.show()
```

* City Fuel Consumption vs. CO2 Emissions: The first plot on the left, shown with blue points, depicts the relationship between fuel consumption in urban (city) driving conditions (in L/100 km) and CO2 emissions (in g/km).
* Highway Fuel Consumption vs. CO2 Emissions: The middle plot with green points shows the relationship between fuel consumption under highway driving conditions (in L/100 km) and CO2 emissions (in g/km).
* Combined Fuel Consumption vs. CO2 Emissions: The plot on the right, with red points, represents the relationship between combined city and highway fuel consumption (in L/100 km) and CO2 emissions (in g/km).
* Each plot demonstrates a positive correlation between fuel consumption and CO2 emissions. The points are more densely packed along an invisible line in each plot, indicating a strong positive relationship. As fuel consumption increases, CO2 emissions also increase across city, highway, and combined conditions. The red plot for combined fuel consumption, in particular, appears to show a slightly tighter correlation, suggesting that combined fuel consumption might be a more consistent predictor of CO2 emissions compared to separate city or highway measurements.

```python
# 2.4 Average CO2 emissions for each car make. Which car make has the highest average and least average impact on CO2 emissions?

plt.figure(figsize=(15, 10))

co2_impact_by_make.sort_values().plot(kind='bar', color='teal')

plt.title('Average CO2 Emissions by Car Make')
plt.xlabel('Car Make')
plt.ylabel('Average CO2 Emissions (g/km)')
plt.xticks(rotation=90)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

* The car make with the highest average impact on CO2 emissions is Bugatti, with an average of 522 grams per kilometer. Conversely, the make with the least impact on CO2 emissions is Smart, with an average of approximately 151.43 grams per kilometer.

* This analysis highlights the significant differences in emissions between luxury, high-performance vehicles and smaller, more efficient city cars

3. **Engine Size and Cylinders:**

```python
# 3.1 What is the relationship between engine size and the number of cylinders?

engine_cylinders_correlation = df['Engine Size(L)'].corr(df['Cylinders'])

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Engine Size(L)', y='Cylinders', data=df)
plt.title('Scatter plot of Engine Size vs. Number of Cylinders')
plt.xlabel('Engine Size (L)')
plt.ylabel('Number of Cylinders')
plt.grid(True)
plt.show()

engine_cylinders_correlation
```

The scatter plot and correlation coefficient indicate a strong positive relationship between engine size and the number of cylinders, with a correlation coefficient of approximately 
0.928. This high correlation suggests that typically, as the engine size increases, so does the number of cylinders in the engine.

**4. Fuel Type Comparison:**

```python
# 4.1 Compare the CO2 emissions and fuel consumption based on fuel type 

df.head(), df.columns
```

```python
emissions_by_fuel_type = df.groupby('Fuel Type').agg({
    'CO2 Emissions(g/km)': 'mean',
    'Fuel Consumption Comb (L/100 km)': 'mean'
}).reset_index()

plt.figure(figsize=(14, 6))

# CO2 Emissions plot
plt.subplot(1, 2, 1)
sns.barplot(x='Fuel Type', y='CO2 Emissions(g/km)', data=emissions_by_fuel_type, palette='viridis')
plt.title('Average CO2 Emissions by Fuel Type')
plt.ylabel('Average CO2 Emissions (g/km)')
plt.xlabel('Fuel Type')

# Fuel Consumption plot
plt.subplot(1, 2, 2)
sns.barplot(x='Fuel Type', y='Fuel Consumption Comb (L/100 km)', data=emissions_by_fuel_type, palette='magma')
plt.title('Average Fuel Consumption by Fuel Type')
plt.ylabel('Average Fuel Consumption (L/100 km)')
plt.xlabel('Fuel Type')

plt.tight_layout()
plt.show()
```

* Average CO2 Emissions by Fuel Type:
The chart shows variation in CO2 emissions across different fuel types, highlighting which fuels are associated with higher emissions.
* Average Fuel Consumption by Fuel Type:
This chart indicates how much fuel is consumed on average (in liters per 100 kilometers) for each fuel type, providing insights into fuel efficiency.

**5. Vehicle Class Analysis:**

```python
# 5.1 Which vehicle class has the best average fuel efficiency?

vehicle_efficiency = df.groupby('Vehicle Class').agg({
    'Fuel Consumption Comb (L/100 km)': 'mean'
}).sort_values(by='Fuel Consumption Comb (L/100 km)')

# Plot Fuel Consumption by Vehicle Class
plt.figure(figsize=(12, 8))
sns.barplot(x=vehicle_efficiency['Fuel Consumption Comb (L/100 km)'], y=vehicle_efficiency.index)
plt.title('Average Fuel Consumption by Vehicle Class')
plt.xlabel('Fuel Consumption (L/100 km)')
plt.ylabel('Vehicle Class')
plt.show()

vehicle_efficiency
```

The Station Wagon - Small class is the most fuel-efficient on average, followed closely by Compact and Mid-Size vehicle classes. Vans (both Cargo and Passenger) show the highest fuel consumption, making them the least fuel-efficient categories in this dataset.
